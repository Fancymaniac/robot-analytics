Calculate the distance between 2 points

```python
import math
print(" Input the x Coordinates one after the other")
x_1 , x_2  = input(),input()
print(" Input the y Coordinates one after the other")
y_1, y_2 = input(),input()
x= math.pow((int(x_2) - int(x_1)),2)
y= math.pow((int(y_2) - int(y_1)),2)
d = math.sqrt(x-y)
print(d)
```
