Palindrome check

```python
s='racecar'
r= s[::-1]
if (r==s): #Using the 'if' condition to compare the strings and return the appropriate values
  print('True')
else:
  print('False')
```

Find Largest Number in List

```python
a = [9,55,-8960,882]
l = a[0]
for i in a:
    if i>=l:
        l=i
print(l)
```
