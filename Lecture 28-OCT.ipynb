```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
# Scatter plot with seaborn
sns.lmplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot with Regression Line")
plt.show()

# Box plot with seaborn
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box plot by Day")
plt.show()
```

Exercise 1: Basic Line and Bar Charts with Matplotlib
Objective: Practice creating simple line and bar charts using Matplotlib.
Instructions:
1.Create a line plot showing the number of visitors to a website over five days.
1.Days: [1, 2, 3, 4, 5]
2.Visitors: [50, 80, 100, 60, 90]
2.Create a bar chart showing sales numbers for three product categories (A, B, C) with values
[100, 150, 200].
Expected Output:
•A line plot showing a trend in visitor numbers.
•A bar chart comparing sales across three product categories.

```python
import matplotlib.pyplot as plt

# Data for the line plot
Days = [1, 2, 3, 4, 5]
Visitors = [50, 80, 100, 60, 90]

# Create a line plot
plt.plot(Days, Visitors, marker="o")
plt.title("Line Plot - Visitors over 5 Days")
plt.xlabel("Days")
plt.ylabel("Number of Visitors")
plt.show()
```

```python
# Data for the bar chart
categories = ['A', 'B', 'C']
sales = [100, 150, 200]

# Create a bar chart
plt.bar(categories, sales)
plt.title("Bar Chart - Sales by Product Category")
plt.xlabel("Product Categories")
plt.ylabel("Sales Numbers")
plt.show()
```

Exercise 3: Creating Subplots for Multi-Plot Visualization
Objective: Learn to create and customize multiple subplots within a single figure.
Instructions:
• Create a 1x2 grid of subplots:
• First subplot: Line plot showing revenue growth over five months.
• Months: [1, 2, 3, 4, 5]
• Revenue: [1000, 1500, 2000, 2500, 3000]
• Second subplot: Bar chart showing profit in the same months.
• Profits: [200, 300, 400, 500, 600]
• Add titles to both subplots and label the axes appropriately.
Expected Output:
• A figure with two subplots: one showing revenue growth and the other comparing monthly
profit.

```python
import matplotlib.pyplot as plt

months = ["January", "Febuary", "March", "April", "May"]
revenue = [1000, 1500, 2000, 2500, 3000]
profits = [200, 300, 400, 500, 600]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(months, revenue, marker='o', color='b')
axes[0].set_title('Revenue Growth Over 5 Months')
axes[0].set_xlabel('Months')
axes[0].set_ylabel('Revenue')

axes[1].bar(months, profits, color='g')
axes[1].set_title('Monthly Profit')
axes[1].set_xlabel('Months')
axes[1].set_ylabel('Profit')

plt.tight_layout()

plt.show()
```
